```python
# i used settings from my investigation to open and read files properly
files = glob.glob('../../project_1_files/inbox/*.csv')



def get_df_from_files():

    files_list = []

    for file in files:
        df = pd.read_csv(file, encoding='windows-1251', sep=';')
        files_list.append(df)
    
    df = pd.concat(files_list, ignore_index=True)
    
    return df
```

```python
df_orders = get_df_from_files()
```

```python
df_orders.head(2)
```

```python
columns = ["Дата",
            "Номер",
            "Сумма",
            "Статус",
            "Примечание",
            "Комментарий",
            "Контрагент",
            "ИНН/КПП",
            "Организация",
            "ИНН/КПП",
            "Тип документа",
            "Имя файла",
            "Дата",
            "Номер 1",
            "Сумма 1",
            "Сумма НДС",
            "Ответственный",
            "Подразделение",
            "Код",
            "Дата",
            "Время",
            "Тип пакета",
            "Идентификатор пакета",
            "Запущено в обработку",
            "Получено контрагентом",
            "Завершено",
            "Увеличение суммы",
            "НДC",
            "Уменьшение суммы",
            "НДС"
            ]
```

```python
df_orders.columns = columns
df_orders.columns = [c.replace(' ', '_') for c in df_orders.columns]
```

```python
df_orders.columns
```

```python
stores = glob.glob('../../project_1_files/Stores/csv/correct/*.csv')

# making a result path
result_path = f'result/{datetime.today().strftime('%Y-%m-%d')}/'
if not os.path.exists(result_path):
    os.makedirs(result_path)


for store in stores:
    df_store = pd.read_csv(store, encoding='Windows-1251', sep=';')
    df_store['Номер счет-фактуры'] = ''
    df_store['Сумма счет-фактуры'] = ''
    df_store['Дата счет-фактуры'] = ''
    df_store['Сравнение дат'] = ''
    # depending on the supplier name i need to add number to the invoice
    # using .iterrows() i can devide index and rows into two objects
    for i, row in df_store.iterrows():

        nnak = row['Номер накладной'] # this will be variable with number
        doc_type = ['СчФктр', 'УпдДоп', 'УпдСчфДоп', 'ЭДОНакл']
        # since i dont know how supplier is named in the df, it is also useful to use 'in' operator
        if 'ЕАПТЕКА' in row['Поставщик']: 
            nnak += '/15'
        
        records = df_orders[df_orders['Номер'] == nnak]
        records = records[records['Тип_документа'].isin(doc_type)]

        if records.empty:
            continue

        invoice = records.iloc[0]['Номер']
        summ = records.iloc[0]['Сумма']
        date = records.iloc[0]['Дата'][1]
        date = datetime.strptime(date, '%d.%m.%y').strftime('%d.%m.%Y')

        df_store.at[i, 'Номер счет-фактуры'] = invoice
        df_store.at[i, 'Сумма счет-фактуры'] = summ
        df_store.at[i, 'Дата счет-фактуры'] = date
        df_store.at[i, 'Сравнение дат'] = '' if (date == df_store.at[i, 'Дата накладной']) else 'Не совпадает!'

    df_store_columns = ['№ п/п', 'Штрих-код партии', 'Наименование товара', 'Поставщик',
        'Дата приходного документа', 'Номер приходного документа',
        'Дата накладной', 'Номер накладной', 'Номер счет-фактуры',
        'Сумма счет-фактуры', 'Кол-во',
        'Сумма в закупочных ценах без НДС', 'Ставка НДС поставщика', 
        'Сумма НДС', 'Сумма в закупочных ценах с НДС', 'Дата счет-фактуры', 'Сравнение дат']


    stores_report = df_store[df_store_columns]

    stores_report.to_excel(f'{result_path}{store.split('/')[6].split('.csv')[0]} - result.xlsx', index=False)
    print(f'{store} done')
```
