```python
# i used settings from my investigation to open and read files properly
files = glob.glob('../../project_1_files/inbox/*.csv')



def get_df_from_files():

    files_list = []

    for file in files:
        df = pd.read_csv(file, encoding='windows-1251', sep=';')
        files_list.append(df)
    
    df = pd.concat(files_list, ignore_index=True)
    
    return df
```

```python
df_orders = get_df_from_files()
```
