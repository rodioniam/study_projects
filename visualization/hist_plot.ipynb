```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from datetime import timedelta
sns.set_style('darkgrid')
```

```python
df_report = pd.read_csv('df_report_conv_daily.csv')
df_report['date'] = pd.to_datetime(df_report['date'])
df_report_mobile = df_report[df_report['deviceCategory'] == 'mobile']
df_report_mobile
```

```python
# using histogram i can see the distribution
plt.figure(figsize=(10, 3))
sns.histplot(df_report_mobile['conv'])
```

```python
# i can customize histogram
# changing bins will give me more precice view
# kde -- this is some kind of trend line
plt.figure(figsize=(10, 3))
sns.histplot(df_report_mobile['conv'], color='black', bins=50, kde=True)
```

```python
df_report
```

```python
# also if im not using filtered dataframe i can use hue
plt.figure(figsize=(10,4))
sns.histplot(df_report, x='conv', hue='deviceCategory', bins=50, kde=True)
```

```python
# if i want to change axies names a can do this
plt.figure(figsize=(10,4))
ax = sns.histplot(df_report, x='conv', hue='deviceCategory', bins=50, kde=True)
ax.set(ylabel='records_count', xlabel='conversion_rate')
```

```python
# i can use line chart to see trends 
# ill use weekly reports for that
plt.figure(figsize=(10,4))
df_report_weekly = pd.read_csv('df_report_conv_weekly.csv')
df_report_weekly['week_date'] = pd.to_datetime(df_report_weekly['week_date'])
ax = sns.lineplot(df_report_weekly, x='week_date', y='conv', hue='deviceCategory')
ax.set(ylabel='conversion_rate')
```
