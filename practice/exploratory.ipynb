```python
import pandas as pd
import seaborn as sns
import kagglehub as kh
from kagglehub import KaggleDatasetAdapter
import matplotlib.pyplot as plt
```

```python
df = kh.dataset_load(KaggleDatasetAdapter.PANDAS,
    handle='uciml/iris',
    path='Iris.csv'
)
```

```python
df.head()
df['Species'].unique()
```

```python
df[df['Species'] == 'Iris-versicolor'].drop('Id', axis=1).describe()
```

```python
df_edited = df.drop('Id', axis=1)
```

```python
sns.pairplot(df_edited)
```

```python
# correlation matrix without a heatmap
df_edited.corr(numeric_only=True)
```

```python
sns.heatmap(df_edited.corr(numeric_only=True), annot=True, vmax=1, vmin=-1)
```

```python
# adding 'style' and 'hue' gives me some kind of clustering
sns.scatterplot(df_edited, x='PetalLengthCm', y='PetalWidthCm', hue='Species', style='Species')
```
