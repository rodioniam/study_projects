```python
import numpy as np
from statistics import mean
from scipy import stats
import pandas as pd
```

```python
arr = np.random.randint(low=1, high=56, size=(23))
arr
```

```python
sum(arr)/len(arr)
```

```python
# here is how i can find mean, first two is from numpy, third is statistics library
arr.mean(), np.mean(arr), mean(arr)
```

```python
# but mean is sensitive to outliers
arr2 = np.insert(arr, 1, 2564)
arr2
```

```python
# here i can clearly see that mean value is much higher then the real distribution is
arr2.mean()
```

```python
# one way to deal with it is to find trimmed mean using stats module from scipy library
# it will cut some min and max values from sample
stats.trim_mean(arr2, proportiontocut=0.2)
# so after cutting 20% of values from sample im getting much closer to the real mean value
```

```python
# other option is to find weighted mean
# it will multiply values by weights and divide by weights sum
# weights can be added by significanse of the values. So i can add less weights to outliers

weights = [] 

for el in arr2:
    if el < 60 and el > 45:
        weights.append(0.6)
    elif el < 30:
        weights.append(0.4)
    else:
        weights.append(0.1)

weights = np.array(weights)
weights

# weights sample must be same size as main sample

# weighted mean is often used in physics and economica
```

```python
# for finding weighted mean i will need to use np.dot()
arr2.dot(weights)
# or
np.dot(arr2, weights)
```

```python
# so the weighted mean equasion will look like this
arr2.dot(weights)/sum(weights)
```

```python
# also for weighted mean i can use this 
np.average(arr2, weights=weights)
```

```python
# also i can use median, because it is good for samples with ouliers too
np.median(arr2)
```

```python
# i can use mode also, it returns the value and how many times it is presented in sample
stats.mode(arr2)
```

```python
# mode is more used with categorical or binary [0, 1] data
bind = [1, 0, 1, 1, 1, 0]

stats.mode(bind)
```
