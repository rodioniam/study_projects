```python
import numpy as np
from statistics import mean
from scipy import stats
```

```python
arr = np.random.randint(low=1, high=56, size=(23))
arr
```

```python
sum(arr)/len(arr)
```

```python
# here is how i can find mean, first two is from numpy, third is statistics library
arr.mean(), np.mean(arr), mean(arr)
```

```python
# but mean is sensitive to outliers
arr2 = np.insert(arr, 1, 2564)
arr2
```

```python
# here i can clearly see that mean value is much higher then the real distribution is
arr2.mean()
```

```python
# one way to deal with it is to find trimmed mean using stats module from scipy library
# it will cut some min and max values from sample
stats.trim_mean(arr2, proportiontocut=0.2)
# so after cutting 20% of values from sample im getting much closer to the real mean value
```
