```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 30)
pd.set_option('display.float_format', '{:.2f}'.format)
df = pd.read_csv('data.txt', encoding='windows-1251')
```

```python
df.head()
```

```python
# here i count how many times each product sold
df.groupby(['DR_NDrugs'])[['DR_Kol']].agg('sum')
sum_by_name = df.groupby(['DR_NDrugs']).agg({'DR_Kol': 'sum'})
sum_by_name
```

```python
# here i find what share of total sales does each product make up
share_df = (sum_by_name.DR_Kol / sum(sum_by_name['DR_Kol'])).sort_values(ascending=False).reset_index()
share_df
```

```python
share_df['cum_sum'] = share_df['DR_Kol'].cumsum()
share_df
```

```python
# this is some kind of if-else operator
share_df['ABC_grp'] = np.where(share_df['cum_sum']< 0.8, 'A', np.where(share_df['cum_sum']< 0.95, 'B', 'C'))
# does work like this: where something something, do this, else do this
share_df
```
