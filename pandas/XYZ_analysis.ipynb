```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 30)
pd.set_option('display.float_format', '{:.2f}'.format)
df = pd.read_csv('data1.txt', encoding='windows-1251')
```

```python
df.head(2)
```

```python
df_g = df.groupby(['DR_Dat', 'DR_NDrugs']).agg({'DR_Kol': 'sum'})
df_g
```

```python
# to calculate standart diviation i need to do this (using std method)
df_g.groupby(['DR_NDrugs']).apply(lambda x: x.std()/x.mean()).reset_index()
# one problem with std method is that if object has only one value it will return NaN
```

```python
# to deal with NaN i can use std from numpy, it will return 0
df_g.groupby(['DR_NDrugs']).apply(lambda x: np.std(x)/x.mean()).reset_index()
# but its better not to have such products in analysis at all
```

```python
# so i will filter my df from this products
df_f = df_g.reset_index().groupby(['DR_NDrugs']).agg({'DR_Dat': 'count'}).reset_index()
xyz_names = list(df_f[df_f['DR_Dat'] > 1]['DR_NDrugs'])
```

```python
df_g = df_g.reset_index()
```

```python
df_g
```

```python
# now i can do same operation, but with filtering
df_g = df_g[df_g['DR_NDrugs'].isin(xyz_names)]
df_g = df_g.groupby(['DR_NDrugs'])[['DR_Kol']].apply(lambda x: x.std()/ x.mean())
df_g
```

```python
df_g['XYZ_grps'] = np.where(df_g['DR_Kol'] < 0.1, 'X', np.where(df_g['DR_Kol'] < 0.25, 'Y', 'Z'))
df_g = df_g.reset_index()
```
