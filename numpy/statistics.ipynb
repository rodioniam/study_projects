```python
import numpy as np
```

```python
a = np.array([[[1, 2, 3], [4, 10, 7]], [[101, 13, 28], [44, 5, 18]]])
```

```python
a.shape, a
```

```python
# the differnce between maximum and minimum values
np.ptp(a)
```

```python
# this function is used when i need to scale values. For example if some values presented as hundreds and others as tens i need to make them somewhat comperrable
# using ptp i can divide element bu ptp value 
# i also can specifie axis
```

```python
# axis=0 will go for each element through all matrices
# axis=1 will go vertically inside each matrix
# axis=2 will go horizontally inside each matrix
np.ptp(a, axis=0), np.ptp(a, axis=1), np.ptp(a, axis=2)
```

```python
# this is function for percentile (item with 20th percentile means that 20% of values are lower/less then this item and 80% are bigger/greater)
# q is the percentile value im looking for
np.percentile(a, q=50, axis=0)
# sometimes percentile value is not present in array, so in interpolates and put number inbetween. This is why some results are float
```

```python
a
```

```python
# i can change interpolation method
# with this array it is not very clear, because i have only to values to choose from, so between 101 and 1 closest is 1
np.percentile(a, q=50, axis=0, method='closest_observation')
```

```python
# there is same function, main difference is in points it calculates
# np.percentile is from 1 to 100 and np.quantile is from 0 to 1
np.quantile(a, q=0.5, axis=0, method='closest_observation')
```

```python
# median, here i cant change interpolation method
np.median(a), np.median(a, axis=1)
```

```python
# average
np.average(a), np.average(a, axis=2)
```

```python
a
```

```python
# variation
np.var(a), np.var(a, axis=1)
```

```python
# also can be used as method
a.var(), a.var(axis=1)
```

```python
# standard deviation
np.std(a),np.std(a, axis=1)
```

```python
# also can be used as method
a.std(), a.std(axis=1)
```
