```python
# if i multiplie arrays (matricies)
c = np.dot(a, b)
c
# it does it this way -- 1st row of a * 1st column of b, then 1st row of a * 2nd column of b  
# [(1*4 + 2*5 + 3*1)=17], [(1*2 + 2*2 + 3*0)=18] and so on
# it is called 'dot product'
```

```python
c.shape
```

```python
# but it is better to use matmul for that
np.matmul(a, b)
# it gives the same result, but algorithm is better suited for that
```

```python
# also np.dot() can be used for multiplying by normal numbers
np.dot(a, 4)
```

```python
# np.matmul() works only with matricies
# also np.matmul() can be replaced with @
a @ b
```

```python
# for multiplying several matricies i can use linalg module from numpy
np.linalg.multi_dot((a, b, a))
```

```python
# also multiplying several matricies can be done by @
a @ b @ a
```

```python
# trace of a matrix - its a sum of all elements of the main diagonal
np.trace(a)
# so for arrray 'a' main diagonal is 1, 5 and sum is 6
```

```python
# also there is one characteristic of a matrix - determinant (det)
# depending on values in the matrix it changes and describes it 
# the matrix should be square 
e = np.array([[2, 2, 2], [1, 1, 1], [4, 4, 4]])
np.linalg.det(e)
```

```python
# this thing helps to find eig values from corr coef matrix
# this values does the greatest impact on the subject, in another words have biggest corr coef in matrix
val, vec =  np.linalg.eig(e)
# this function gives two arrays 
```

```python
# first is eigen values
val
```

```python
# second is eigen vectors
vec
```
