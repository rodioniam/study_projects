```python
# Inserting values
```

```python
b ,b.shape
```

```python
# obj - indicates where to insert by index
# values - what to insert
# axis - on what axis to insert
np.insert(b, obj=1, values=404, axis=1)
```

```python
# obj and values can read collections of values
# if im using both multiple obj and values each index corresponds to a specific value (in order)
np.insert(b, obj=(1, 2), values=(404, 6666), axis=1)
```

```python
# but if i dont put any axies my n-dimensional array will become 1d
np.insert(b,(1,2) ,(404, 606))
```

```python
# i also can use append for arrays
# dimension of appending values must be the same as the dimension of array, in this case i have size (2, 3), so 2 
np.append(b, values=[[404], [606]], axis=1)
```

```python
# this is how i can work with unique values in array
b1 = np.append(b, [[1, 2, 3]], axis=0)
b1 = np.append(b1, [[1], [1], [1]], axis=1)
b1
```

```python
# 0 is vertically
# 1 is horizontally
np.unique(b1, axis=0), np.unique(b1, axis=1)
```

```python
# return_index if True returns indexes of unique elements in array
np.unique(b1, axis=0, return_index=True)
```

```python
# return_inverse if True returns indexes in order so we know to recreate initial array
# in this case i need to take 0 (first) element, then 1 (second) and again 0 (first)
np.unique(b1, axis=0, return_inverse=True)
```

```python
# return_counts if True returns how many times each element repeates
np.unique(b1, axis=0, return_counts=True)
```

```python
# filtering arrays 
```

```python
o = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
o
```

```python
# to filter array i can use np.where()
np.where(o < 4)
# returns two arrays 
# first is indexes of rows where elements where found
# second one is indexes of columns of elements
```

```python
np.where(o%2 == 1)
```

```python
# with this i can get values themselfs
o[np.where(o%2 == 1)]
```

```python
# if i want to use more then one condition
np.where((a%2==1) & (a>3))
#
o[np.where((a%2==1) & (a>3))] # this is like using 'and' in if statment
#
o[np.where((a%2==1) | (a>3))] # this is like using 'or' in if statment
```

```python
# i also can create if-else
# it reads like np.where(if a > 5, return a, else return -1)
np.where(a > 5, a, -1)
```
