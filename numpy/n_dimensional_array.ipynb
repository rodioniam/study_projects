```python
arr3 = [
    [[1, 2, 3], [4, 5, 6]],
    [[1, 2, 3], [4, 5, 6]],
    [[1, 2, 3], [4, 5, 6]]
]

c = np.array(arr3)
```

```python
c
```

```python
c.shape
# 3 inside lists, 2 in each of them, 3 elements in each of 2
```

```python
c.ndim
```

```python
# math operations are the same
```

```python
# agg operations
c.mean()
```

```python
c
```

```python
# it is going like deeper/thru the tables
c.mean(axis=0)
```

```python
# it is vertically
c.mean(axis=1)
```

```python
# it is horizontally
c.mean(axis=2)
```

```python
# with nD arrays i can put multiple axies
c.mean(axis=(0,1))
```

```python
# this way i can create standart 2d matrix with 1 values on diagonal and all rest are 0
np.eye(10, 10) # first arg is N - number of rows, second is M - number of columns
# M is optional, if not given it defaults to N value
```

```python
# this way i can create custom array with 0 or 1 in it
np.zeros(shape=(2, 3, 5)) # in shape i type in what shape for array i want
np.ones(shape=(2, 3, 5))
```

```python
# creates new array with the shape of existing one and fills it with 1
np.ones_like(arr)
```

```python
# does the same, but does not put 0 in array
np.empty(shape=(2, 3, 5))
# use with caution!
```

```python
# also creates custom array, but lets me fill with desired value
np.full(shape=(2, 3, 5), fill_value=808)
```

```python
# if i want to make fractional range i cant use range(), since it cant accept float type
# instead i can use numpy
np.arange(0, 10, 0.1)
```

```python
# in case when i know start and end value and need exact number of values in array i can use this
np.linspace(0, 10, 5) # i need range of 5 values from 0 to 10 
```
