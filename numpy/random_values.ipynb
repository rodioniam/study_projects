```python
import numpy as np
```

```python
a = [1, 2, 3, 4, 5, 6]
# creates array from random values from collection
np.random.choice(a, size=(3,3,2))
```

```python
# it can also generate from a number. It will generate a range from 0 to this number
np.random.choice(10, (2, 2, 4))
```

```python
# there is replace argument that determines whether there can be duplicated values
# as a default np.random.choice duplicates values
np.random.choice(10, (2, 2, 2), replace=False)
```

```python
# there is probability argument -- p
p = [0, 0.3, 0, 0, 0, 0.7]
a, p
```

```python
# it gives a probability for values to appear in generated array
# p must be the same size as collection and probabilities must sum to 1
np.random.choice(a, (3, 4), p=p)
```

```python
b = np.array([1, 2, 3, 4, 5])
# shuffles the given array
np.random.shuffle(b)
```

```python
# it does it in place (changes variable)
b
```

```python
# there is a similar functiot that shuffles, but not in place
b = np.array([1, 2, 3, 4, 5])
np.random.permutation(b)
```

```python
# np.random.permutation() can also work like choice if give it only number
np.random.permutation(20) 
```

```python
# generates random value in range from 0 to 1 
np.random.sample()
```

```python
# can also generate array with random values
np.random.sample((2, 3, 5))
```

```python
# generates array with random values (0-1)
np.random.rand(2, 3, 5)
```

```python
# generates random integer within a given range
np.random.randint(low=10, high=25)
```

```python
# can also generate an array
np.random.randint(10, 25, (2, 3, 4))
```

```python
# also in numpy i can choose what tyoe of distribution i need
np.random.normal(5, 1, 10)
# first is expected value, second is dispersion and last one is size
```

```python
np.random.uniform(5, 10, 10)
```
