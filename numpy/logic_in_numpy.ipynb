```python
import numpy as np
```

```python
a = np.array([[1, 2, 3, np.nan], [4, np.inf, 5, -np.inf]])
b = np.array([[1, 2, 3, np.nan, True], [4, np.inf, 5, -np.inf, False]])
```

```python
# checks if all elements of array are present, or can be interpreted as True
np.all(a)
```

```python
np.all(b)
```

```python
b
```

```python
# also works with axis
np.all(b, axis=1), np.all(b, axis=0)
```

```python
# np.any() checks are there any elements in array that can be interpreted as true
np.any(b, axis=1)
```

```python
c = np.array([[1, 2, 3, np.nan, True], [4, np.inf, 5, -np.inf, False], [False, False, 0, False, False]])
```

```python
np.any(c, axis=1)
```

```python
b
```

```python
# checks if the elements in array are infinite
np.isfinite(b)
```

```python
# does the opposite
np.isinf(b)
# notice thet nan is considered neither infinite or finite
```

```python
# cheks if element is nan
np.isnan(b)
```

```python
# there is a function to check how close are the values in different arrays
d = np.array([0.000005, 0.0003, 0.008])
e = d + 10000
e
```

```python
# atol -- absolute tolerance
np.isclose(d, e, atol=9999.9)
```

```python
# does the same, returns result for all elements
np.allclose(d, e, atol=9999.9)
```

```python
aa = a.copy()
a.shape, aa.shape
```

```python
# compare arrays, must be same shape
aa == a
# or 
np.equal(aa, a)
```

```python
# does the same but for all els at once
# equal_nan=True -- means that nans in arrays are considered equal
np.array_equal(a, aa, equal_nan=True)
```
