```python
import numpy as np
```

```python
# one-dimensional array
d1 = np.array([1, 2, 3, 4, 5])
d1
```

```python
# everything in numpy works with this type
type(d1)
```

```python
lst = ['a', 'b', 'c']

d11 = np.array(lst)
```

```python
# basic things with arrays
d1.shape # returns number of elements
d1.ndim # shows number of dimensions
d1.dtype # returns type of the array, type of the elements in it
```

```python
# this way i can set dtype manually
d1 = np.array([1, 2, 3, 4, 5], dtype=np.float64)
```

```python
# this way i can change dtype of already created array
d1.astype(np.int32)
```

```python
# another way of creating array
lst2 = [1, 2, 3, 4, 5]

np.asarray(lst2) 
```

```python
# summing np.arrays will gave array with results of summing like if was for loop
a = np.array([1, 2, 3, 4])
b = np.array([4, 5, 6, 7])

a + b 

# this works with all math operations like -, *, /
# when dividing by 0 with numpy it will give 'inf', not an error
```

```python
# i can aggregate arrays
a.sum()
```

```python
a.mean()
```
